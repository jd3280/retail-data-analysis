任务 1.5
统计每位顾客每月的消费额及消费天数，并在报告中列出用户编号为 0-10 的结果 。


```python
import pandas as pd
sale = pd.read_csv('/Users/dingjiadong/Desktop/uber/task1_1.csv')
```

```python
mo_sale = sale[['顾客编号','销售日期','销售金额']]
```

```python
mo_sale['销售日期'] = pd.to_datetime(sale['销售日期'], format='%Y%m%d')
mo_sale.set_index(mo_sale['销售日期'],inplace=True)
```

```python
mo_cust = mo_sale.groupby(['顾客编号',pd.Grouper(freq='M')]).sum()
```

```python
mo_cust
```

```python
mth_sale = sale[['顾客编号','销售日期','销售月份','销售金额',]]
mth_sale
```

```python
#mth_sale.loc[mth_sale.顾客编号==1].销售日期
```

```python
mth = [201501,201502,201503,201504]


sale_day = []   
for i in range(0,2612):
    for m in mth: 
        day = len(mth_sale.loc[mth_sale.顾客编号==i].loc[mth_sale.销售月份==m].销售日期.unique())

        sale_day.append(day)


```

```python
#len(mth_sale.loc[mth_sale.顾客编号==0].loc[mth_sale.销售月份==201501].销售日期.unique())
```

```python
sale_day = pd.DataFrame(sale_day, columns=['消费天数'])
sale_day.head(10)
```

```python
sale_day = sale_day.drop(index=sale_day.消费天数[sale_day.消费天数==0].index)
```

```python
len(sale_day)
```

```python
len(list(sale_day.消费天数))
```

```python
mo_cust['消费天数'] = list(sale_day.消费天数)
```
