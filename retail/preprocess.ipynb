1 数据预处理与统计

```python
import pandas as pd
sale = pd.read_csv('/Users/dingjiadong/Desktop/uber/data.csv',encoding='GB18030')
```

```python
sale.head()
```

查看数据集是否存在数据缺失，数据重复，并且进行删除。

```python
sale.isna().sum()
```

```python
sale.dropna(inplace=True)
sale.shape
```

```python
sale.drop_duplicates(inplace=True)
sale.shape
```

在后续操作中发现日期存在异常，查看错误日期数据，并将错误日期数据移到下一天。

```python
sale[sale['销售日期']==20150229]
```

```python
sale.loc[22249,'销售日期'] = 20150301
sale.loc[22250,'销售日期'] = 20150301
```

```python
sale[sale['销售日期']==20150301]
```

将销售金额与促销不匹配的进行修正

```python
idx = sale[sale['销售金额'] != sale['销售数量']*sale['商品单价']].index
```

```python
idx
```

```python
sale.loc[idx,'是否促销']='是'
```

将销售金额，数量，单价进行规范

```python
sale = sale.loc[(sale.销售数量) >0 & (sale.销售金额 >=0) & (sale.商品单价 >0)]
```

```python
sale.to_csv('/Users/dingjiadong/Desktop/uber/task1_1.csv')
```
