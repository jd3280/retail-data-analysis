```python
import pandas as pd
sale = pd.read_csv('/Users/dingjiadong/Desktop/uber/task1_1.csv')
```

```python
big_class = sale[['大类名称','销售月份','销售金额']]
```

```python
big_class.groupby(by=['销售月份','大类名称']).sum()
```

```python
jan = big_class.loc[big_class.销售月份==201501].groupby('大类名称').sum()
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']

```

```python
plt.figure(figsize=(12,12))

plt.pie(jan['销售金额'], labels=jan.index,autopct='%1.1f%%')
plt.title('一月各大类商品销售额占比')
plt.show()
```
